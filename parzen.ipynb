```
import numpy as np
import matplotlib.pyplot as plt
```

```

d = np.array([-7, -5, -4, -3, -2, 0, 2, 3, 4, 5, 7])
d = np.repeat(d[np.newaxis, :], x.shape[0], axis=0)
x = np.linspace(-8, 8, 1000)
x = np.repeat(x.T[:, np.newaxis], d.shape[1], axis=1)
fig, axs = plt.subplots(1, 3)
fig.set_size_inches(20, 4)

js = [1, 4, 11]

for cnt, j in enumerate(js):
  h = 1/np.sqrt(j)
  k = (x - d) / h
  k = np.apply_along_axis(lambda  t: 1*(np.abs(t) <= 0.5), axis=1, arr=k)
  k = k.mean(axis=1) / h

  axs[cnt].plot(x, k)
  axs[cnt].set_title('j={}'.format(j))

plt.show()

```

The bigger the j is, the smaller the h becomes therefore the estimated distributaion would be spikier

```
d = np.array([-7, -5, -4, -3, -2, 0, 2, 3, 4, 5, 7])
d = np.repeat(d[np.newaxis, :], x.shape[0], axis=0)
x = np.linspace(-8, 8, 1000)
x = np.repeat(x.T[:, np.newaxis], d.shape[1], axis=1)


js = [1, 2, 4, 10]
fig, axs = plt.subplots(1, len(js))
fig.set_size_inches(20, 4)

for cnt, j in enumerate(js):
  h = j
  k = (x - d) / h
  k = np.apply_along_axis(lambda  t: 1*(np.abs(t) <= 0.5), axis=1, arr=k)
  k = k.mean(axis=1) / h

  axs[cnt].plot(x, k)
  axs[cnt].set_title('j={}'.format(j))

plt.show()
```

The bigger the h is, the smoother the distribution becomes because the h would be bigger
